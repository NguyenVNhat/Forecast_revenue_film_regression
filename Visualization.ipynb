```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data.csv')
```

```python
data.tail(10)
```

```python
fig,axe = plt.subplots(figsize = (14,5))
tempdata = data.tail(20)
x = tempdata['Maxtheater']
y = tempdata['Gross']

sns.barplot(data=tempdata,x= x,y =y)
```

Mô tả : nhìn theo sơ đồ bar ta thấy được sự tương quan của theater và gross
- Với theater càng lớn thì gần như là gross càng lớn

```python
fig,axe = plt.subplots(figsize = (14,5))
sns.barplot(data=data,x= data['Year'],y =data['Gross'])
```
